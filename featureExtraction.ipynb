### 用于编写提取DNA序列特征的代码

```python
import itertools
import matplotlib.pyplot as plt
from tqdm import tqdm
import pandas as pd
import sys
from Bio import SeqIO
import itertools
import os
from utils import featureExtraction
import seaborn as sns
from sklearn.decomposition import PCA

from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import networkx as nx

from sklearn.manifold import LocallyLinearEmbedding

from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
data_path="data/S1.csv"
save_path="data/S1_feature/S1_feature_nolzcurve.csv"
save_path_lz="data/S1_feature/lzcurver/"
save_path_extraction="data/S1_feature/featureExtraction/"
```

```python
data=pd.read_csv(data_path)
seqs=list(data['nucleotide'])
fseqs=[featureExtraction.get_reverse_seq(s) for s in seqs]
# labels=list(data['label'])
```

```python
GCContent=pd.DataFrame([featureExtraction.get_feature_GCContent(s)[0] for s in seqs],columns=["GC Content"])
zcurver=pd.DataFrame([featureExtraction.get_feature_zCurve(s) for s in seqs],columns=["zCurveX","zCurveY","zCurveZ"])
cumulativeSkew=pd.DataFrame([featureExtraction.get_feature_cumulativeSkew(s) for s in seqs],columns=["GC Skew","AT Skew"])
atgcRatio=pd.DataFrame([featureExtraction.get_feature_atgcRatio(s)[0] for s in seqs],columns=["ATGC Ratio"])
```

```python
kmer1=pd.DataFrame([featureExtraction.get_feature_kmer(s,1) for s in seqs],columns=["kmer:"+name for name in featureExtraction.get_kmer(1)])
kmer2=pd.DataFrame([featureExtraction.get_feature_kmer(s,2) for s in seqs],columns=["kmer:"+name for name in featureExtraction.get_kmer(2)])
kmer3=pd.DataFrame([featureExtraction.get_feature_kmer(s,3) for s in seqs],columns=["kmer:"+name for name in featureExtraction.get_kmer(3)])

# rkmer1=pd.DataFrame([featureExtraction.get_feature_kmer(s,1) for s in fseqs],columns=["rkmer:"+name for name in featureExtraction.get_kmer(1)])
# rkmer2=pd.DataFrame([featureExtraction.get_feature_kmer(s,2) for s in fseqs],columns=["rkmer:"+name for name in featureExtraction.get_kmer(2)])
# rkmer3=pd.DataFrame([featureExtraction.get_feature_kmer(s,3) for s in fseqs],columns=["rkmer:"+name for name in featureExtraction.get_kmer(3)])
```

```python
# col=featureExtraction.get_kmer(1)+featureExtraction.get_kmer(2)+featureExtraction.get_kmer(3)
pseknc=pd.DataFrame([featureExtraction.get_feature_pseKnc(s,3) for s in seqs],columns=["pseknc:"+str(name) for name in range(1,66)])
pseknc
```

```python
# feature=pd.concat([GCContent,zcurver,cumulativeSkew,atgcRatio,kmer1,kmer2,kmer3,rkmer1,rkmer2,rkmer3,pseknc,data['label']],axis=1)
feature=pd.concat([GCContent,zcurver,cumulativeSkew,atgcRatio,kmer1,kmer2,kmer3,pseknc,data['label']],axis=1)
feature.to_csv(save_path,index=False)
```

```python
for w in range(1,7):
    for k in range(0,7):
        file_path = save_path_lz + "zcurve_w" + str(w) + "k" + str(k) + ".csv"
        if not os.path.exists(file_path):
            lz=featureExtraction.get_feature_lzcurver(seqs,w,k,1)
            lz=pd.DataFrame(lz,columns=featureExtraction.lz_col_name(w,k,1))
            lz.to_csv(save_path_lz+"zcurve_w"+str(w)+"k"+str(k)+".csv",index=False)
```
